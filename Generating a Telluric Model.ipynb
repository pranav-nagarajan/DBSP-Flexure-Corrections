```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14

from astropy.convolution import Gaussian1DKernel, convolve

import warnings
warnings.simplefilter('ignore')

from telfit import Modeler
```

```python
def convert_air_to_vacuum_wl(wave_air):
    '''Only valid above 2000 Angstroms.'''
    wave_vac = np.copy(wave_air)
    for i in [0, 1]:
        sigma2 = (1e4 / wave_vac) ** 2
        f = 1.0 + 0.05792105 / (238.0185 - sigma2) + 0.00167917 / (57.362 - sigma2)
        wave_vac = wave_air * f
    return wave_vac
```

```python
# Example script to generate a template spectrum for telluric absorption lines
wave_feros = np.arange(5000, 10000, 0.02)
    
# Palomar Observatory
alt = 1.7
lat = 33.4
R = 25000
    
def run_telfit(h2o, airmass):
 
    # Start / end wavelengths in nm
    wavestart = 500.0
    waveend = 1000.0
           
    # Generate telluric spectrum with default resolution 0.001 Angstroms
    modeler = Modeler()
    angl = np.degrees(np.arccos(1./airmass))

    model = modeler.MakeModel(humidity=h2o, angle=angl, lowfreq=1e7/waveend, highfreq=1e7/wavestart, lat=lat, alt=alt)
    tell = model.toarray()

    # Convert to Angstroms
    wave = 10. * tell[:,0]
    flux = tell[:,1]
    
    # Determine binning
    dl = np.median(np.diff(wave))
    
    # Smooth and rebin to linear 0.02 Angstrom pixels
    sigma_aa = 7600/R/2.35
    sig_res = sigma_aa/dl
    print('Kernel ',sig_res)
    
    # Smoothen the telluric spectrum
    gauss_kernel = Gaussian1DKernel(sig_res)
    smooth_flux = convolve(flux, gauss_kernel)
    tell_flux = np.interp(wave_feros, wave, smooth_flux)
    return tell_flux

# Compute telluric model for grid of airmasses
a_grid = np.arange(1.0, 2.05, 0.05)
humidity_guess = 35
all_tells = [run_telfit(h2o = humidity_guess, airmass = ai) for ai in a_grid]

# Convert to wavelengths in vacuum for consistency
wl_vac = convert_air_to_vacuum_wl(wave_feros)
np.savez('telluric_grid_airmass_r25000.npz', wl_grid = wl_vac, airmass_vals = a_grid, telluric_grids = all_tells)
```
